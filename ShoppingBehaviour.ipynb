```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/shopping_behavior_updated.csv')
```

```python
df.head(30)
```

Making New Column(age_group),through Given Age Column

```python
df.loc[df['Age']<=31, 'age_group'] = 'Young'
df.loc[df['Age'].between(32,44), 'age_group'] = 'Adult'
df.loc[df['Age'].between(45,57), 'age_group'] = 'Old'
df.loc[df['Age']>=70, 'age_group'] = 'Very_Old'
# df.insert(4,'age_group',df['Age'],allow_duplicates = True)
```

```python
df.info()
```

```python
df.describe()
```

Removing Un-necessary Columns

```python
df.drop(columns=['Payment Method','Shipping Type'],inplace=True)
```

Performing Exploratory Data Analysis :-

```python
df.head()
```

```python
Category_Wise= df.groupby('Category')['Purchase Amount (USD)'].sum().sort_values(ascending=True)
plt.plot(Category_Wise.index,Category_Wise,marker='o')
plt.title("Top-5 Category Item Purchased")
```

```python
Item_Wise= df.groupby('Item Purchased')['Purchase Amount (USD)'].sum().sort_values(ascending=True).head()
plt.plot(Item_Wise.index,Item_Wise,marker='o')
plt.title("Top-5 Item Purchased")
# plt.xlabel("Average Pulse")
# plt.ylabel("Count Of Item")
```

```python
Age_Group=df.groupby('age_group')['Purchase Amount (USD)'].sum().sort_values(ascending=True).head()
plt.barh(Age_Group.index,Age_Group)
plt.title("Age Group Purchase")
```

```python
Gender_Wise=df.groupby('Gender')['Purchase Amount (USD)'].sum().sort_values(ascending=True).head()
# plt.pie(Gender_Wise)
ax =Gender_Wise.plot(kind="pie", autopct='%1.1f%%',legend=True, title='Gender V/S Purchase', ylabel='', labeldistance=None)
ax.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
plt.show()
```

```python
Size_Wise=df.groupby('Size')['Purchase Amount (USD)'].count().sort_values(ascending=True).head()
plt.bar(Size_Wise.index,Size_Wise)
plt.title("Size v/s Size Sold")

```

```python
Season_Wise=df.groupby('Season')['Purchase Amount (USD)'].count().sort_values(ascending=True).head()
# plt.pie(Season_Wise)
# df.pie(Season_Wise,labels=labels, autopct='%.1f%%')
# plt.set_title('Sport Popularity')
ax =Season_Wise.plot(kind="pie", autopct='%1.1f%%',legend=True, title='Season Wise Purchase Rate', ylabel='', labeldistance=None)
ax.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
plt.show()
```
